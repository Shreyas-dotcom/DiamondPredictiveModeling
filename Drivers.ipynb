```python
import pandas as pd
import statsmodels.api as sm

# Load the dataset
df = pd.read_csv('testing_data.csv')

# Prepare the data for regression analysis
X = df[['carat', 'cut', 'color', 'clarity', 'depth', 'table']]
X = pd.get_dummies(X, drop_first=True)  # Convert categorical variables to dummy variables
y = df['price']

# Add a constant column to the independent variables
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Get the key drivers (coefficients) and their p-values
drivers = pd.DataFrame({'Feature': X.columns, 'Coefficient': model.params, 'P-value': model.pvalues})

# Print the key drivers
print(drivers)
```

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('testing_data.csv')

# Prepare the data for regression analysis
X = df[['carat', 'cut', 'color', 'clarity', 'depth', 'table']]
X = pd.get_dummies(X, drop_first=True)  # Convert categorical variables to dummy variables
y = df['price']

# Add a constant column to the independent variables
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Get the key drivers (coefficients) and their p-values
drivers = pd.DataFrame({'Feature': X.columns, 'Coefficient': model.params, 'P-value': model.pvalues})

# Sort the drivers by coefficient values
drivers = drivers.sort_values(by='Coefficient', ascending=False)

# Plot the bar chart
plt.figure(figsize=(10, 6))
plt.bar(drivers['Feature'], drivers['Coefficient'])
plt.xlabel('Feature')
plt.ylabel('Coefficient')
plt.title('Key Drivers')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('testing_data.csv')

# Prepare the data for regression analysis
X = df[['carat', 'cut', 'color', 'clarity', 'depth', 'table']]
X = pd.get_dummies(X, drop_first=True)  # Convert categorical variables to dummy variables
y = df['price']

# Add a constant column to the independent variables
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Get the key drivers (coefficients) and their p-values
drivers = pd.DataFrame({'Feature': X.columns, 'Coefficient': model.params, 'P-value': model.pvalues})

# Sort the drivers by coefficient values
drivers = drivers.sort_values(by='Coefficient', ascending=False)

# Plot the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(drivers['Feature'], drivers['Coefficient'])
plt.xlabel('Feature')
plt.ylabel('Coefficient')
plt.title('Key Drivers')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('diamonds.csv')
```

```python
missing_values = df.isnull().sum()
data_quality_score = 1 - (missing_values / len(df))
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(data_quality_score.to_frame(), cmap='YlGnBu', annot=True, cbar=False)
plt.title('Data Quality Scores')
plt.ylabel('Features')
plt.xlabel('  Score')
plt.show()
```
